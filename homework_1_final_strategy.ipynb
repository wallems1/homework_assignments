1) Our 1st strategy bases its move selection on randomly generating a first move and then successively picking the next move based on that. Essentially, it starts a sequence of moves (ie. R-P-S-R-P-S-...) and keeps that sequence of moves throughout play time. This strategy is simple and not very dynamic.

2) Our 2nd strategy relies on a few different things, primairily on the amount of each moves taken by our opponent.
Essentially, counters are initialized to keep track of each rock, paper and scissors move taken and are kept track of
when partner history is updated.

In the cases where one move is found to be played less than the others, we select the move to play against the opponent that would beat them (ie. if rock is being played the least, we play the move that would defeat it (paper)). In the cases where two move type amounts are equal and less than the 3rd option, we select the move that would defeat that 3rd option. In the final edge case where all move counts are equal, we choose randomly a move.

Additionally, our model considers what happens when fitness is poor. If our fitness is under some threshold (after some amount of moves) we throw a random move to reset our trajectory.

```python
class Vengeance:
    def __init__(self):
        self.name = "Vengeance"
        self.moves = ['R', 'P', 'S']
        self.my_history = []
        self.partner_history = []
        self.fitness = 0
        self.countR = 0
        self.countP = 0
        self.countS = 0
        
    def reset(self):
        self.my_history = []
        self.partner_history = []
        self.fitness = 0
        
    def play(self):
        # if play history is empty, initialize randomly a move from self.moves
        if len(self.my_history) == 0:
            my_move = np.random.choice(self.moves)
            self.my_history.append(my_move)
            
        # check number of moves made and current fitness
        # if conditions met, reset with a random move
        if len(self.my_history) > 10 and self.fitness < 5:
            my_move = np.random.choice(['R','S','P']) 
        
        # if R counter is least, then throw P to beat R
        if self.countR < self.countS and self.countR < self.countP:
            my_move = 'P'
        # if P counter is least, then throw S to beat P
        if self.countP < self.countS and self.countP < self.countR:
            my_move = 'S'   
        # if S counter is least, then throw R to beat S
        if self.countS < self.countP and self.countS < self.countR:
            my_move = 'R'
            
        # if R and P counts equal and less than S, play R to beat S
        if self.countR == self.countP and self.countR < self.countS:
            my_move = np.random.choice(['R'])
            
        # if P and S counts equal and less than R, play P to beat R
        if self.countP == self.countS and self.countP < self.countR:
            my_move = np.random.choice(['P'])
            
        # if S and R counts equal and less than P, play S to beat P
        if self.countS == self.countR and self.countS < self.countP:
            my_move = np.random.choice(['S']) 
            
        # if S and R and P counts equal, pick randomly
        if self.countS == self.countR and self.countS == self.countP:
            my_move = np.random.choice(['R','S','P']) 
            
        self.my_history.append(my_move)
        return(my_move)
        
    def update_partner_history(self, their_move):
        # count R moves
        if their_move == 'R':
            self.countR = self.countR + 1
        # count S moves
        if their_move == 'S':
            self.countS = self.countS + 1
        # count P moves
        if their_move == 'P':
            self.countP = self.countP + 1
        self.partner_history.append(their_move)
    
    def replicate(self):
        offspring = type(self)()
        return(offspring)
```
